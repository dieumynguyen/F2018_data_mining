(b)  Generate a scatter plot using the T2 and T4 attributes. You can use any plotting tool, such as excel, matlab, gnuplot, R, python, etc. 

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# Read csv
df = pd.read_csv("data/dataset.csv")
```

```python
# Plot scatter & save pdf
x = df['T2']
y = df['T4']

f = plt.figure()
plt.scatter(x, y, marker='o', alpha=0.3, color="#a1c3d2");
plt.xlabel('T2 (Celsius)')
plt.ylabel('T4 (Celsius)')
plt.title('Temperature in living room area (T2) vs Temperature in office room (T4)');
f.savefig('Nguyen_DieuMy_HW1_Q3b_figure.pdf', bbox_inches='tight')
```
